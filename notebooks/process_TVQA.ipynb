### Modifying the code samples and manually executing them

```python
from main_simple_lib import *
```

```python
import decord
from decord import cpu, gpu
import numpy as np

def my_get_video(video_path):
    # CTRL C/V from internal code
    video_reader = decord.VideoReader(str(video_path), num_threads=1, ctx=cpu(0))
    decord.bridge.set_bridge('torch')
    vlen = len(video_reader)
    original_fps = video_reader.get_avg_fps()
    num_frames = int(vlen * 30 / original_fps)
    num_frames = min(30, num_frames)
    frame_idxs = np.linspace(0, vlen, num_frames, endpoint=False).astype(int)
    video = video_reader.get_batch(frame_idxs).byte()
    video = video.permute(0, 3, 1, 2)
    return video

video = my_get_video("data/videos/house_s07e01_seg02_clip_12.mp4")
```

#### Messing with first prompt output (allowed to use `llm_query()` method)

```python
frame_info = []
def modified_execute_command(video):
    video_segment = VideoSegment(video)
    for frame in video_segment.frame_iterator():
        exists_wilson = frame.exists("Wilson")
        exists_house = frame.exists("House")
        query_breaking = frame.simple_query("What room is Wilson breaking into?")
        frame_info.append((exists_wilson, exists_house, query_breaking))
        if exists_wilson and exists_house and query_breaking == "yes":
            return frame.llm_query("What room was Wilson breaking into when House found him?")
    
modified_execute_command(video)
```

```python
# ID all the frames that could have been correct answers
# ie: where exists(wilson) and exists(house) returns True.
important_frames = [(frame_idx, qB) for frame_idx, (eW, eH, qB) in enumerate(frame_info) if eW and eH]

import matplotlib.pyplot as plt
# plot the frame before, the liftoff frame, and the frame after
fig, ax = plt.subplots(len(important_frames), 1, figsize=(5,5 * len(important_frames)))


video_segment = VideoSegment(video)
for i, (frame_idx, query_breaking_val) in enumerate(important_frames):
    ax[i].imshow(video[frame_idx].permute(1,2,0))
    ax[i].set_title(f"Frame {frame_idx} - query(What room is Wilson breaking into) = {query_breaking_val}")
    executable_frame = video_segment.frame_from_index(frame_idx)
    print(f'Frame {i} : {executable_frame.llm_query("What room was Wilson breaking into when House found him?")}')
```
